```python
import pandas as pd

df = pd.read_csv('Data/arc.csv')

df = df.drop(['name', 'version', 'name.1'], axis=1)

df.describe()
```

```python
zeroes = 0
ones = 0

for i in range(len(df['bug'])):
    if df['bug'][i] != 0:  ones += 1
    else: zeroes += 1
    
print('zeros: ', zeroes, ' | Ones: ', ones)
```
